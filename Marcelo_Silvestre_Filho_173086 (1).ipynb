# RMS Titanic

O RMS Titanic foi um navio de passageiros britânico operado pela White Star Line e construído pelos estaleiros da Harland and Wolff em Belfast. Foi a segunda embarcação da Classe Olympic de transatlânticos depois do RMS Olympic e seguido pelo HMHS Britannic. Projetado pelos engenheiros navais Alexander Carlisle e Thomas Andrews, sua construção começou em março de 1909 e ele foi lançado ao mar em maio de 1911. O Titanic foi pensado para ser o navio mais luxuoso e mais seguro de sua época, gerando lendas que era supostamente "inafundável".

A embarcação partiu em sua viagem inaugural de Southampton para Nova Iorque em 10 de abril de 1912, no caminho passando em Cherbourg-Octeville na França e por Queenstown na Irlanda. Ele colidiu com um iceberg às 23h40min do dia 14 de abril e afundou na madrugada do dia seguinte matando mais de 1500 pessoas, sendo um dos maiores desastres marítimos em tempos de paz de toda a história. Seu naufrágio destacou vários pontos fracos de seu projeto, deficiências nos procedimentos de evacuação de emergência e falhas nas regulamentações marítimas da época. Comissões de inquérito foram instauradas nos Estados Unidos e no Reino Unido, levando a mudanças nas leis internacionais de navegação que permanecem em vigor mais de um século depois.

Os destroços do Titanic foram procurados por décadas até serem encontrados em 1985 por uma equipe liderada por Robert Ballard. Ele se encontra a 3843 m de profundidade e a 650 km ao sudeste de Terra Nova no Canadá. Sua história e naufrágio permaneceram no imaginário popular durante décadas, levando a produção de vários livros e filmes a seu respeito, mais notavelmente o filme Titanic de 1997. Até hoje o Titanic permanece como um dos navios mais famosos da história, com seus destroços atraindo várias expedições de exploração ao longo dos anos.

6 - Renomear as colunas restantes para a lingua portuguesa, utilizando os seguintes nomes de colunas: IdPassageiro, Sobreviveu, Classe, Nome, Sexo, Idade, Tarifa, Cabine e Embarque.

7 - Alterar o conteudo da coluna Sobreviveu para:
- 0 => Não
- 1 => Sim

```python
#Substitui, através da função replace na coluna 'Sobreviveu', os valores de 1 e 0, pelos valores 'Sim' e 'Não'
data['Sobreviveu']= data['Sobreviveu'].replace([1,0],['Sim','Não'])

#Exibe as primeiras linhas do DataFrame
data.head(10)
```

8 - Alterar o conteudo da coluna Sexo para:
- female => Mulher
- male   => Homem

```python
#Substitui, através da função replace na coluna 'Sexo', os valores 'male' e 'female, pelos valores 'Homem' e 'Mulher'
data['Sexo']= data['Sexo'].replace(['female','male'],['Mulher','Homem'])

#Exibe as primeiras linhas do DataFrame
data.head(10)
```

10 - Quantos passageiros sobreviveram e quantos não sobreviveram?

```python
#Agrupa o DataFrame pela coluna 'Sobreviveu' e conta as quantidades de ocorrencias
sobreviventes = data.groupby('Sobreviveu')['Sobreviveu'].count().reset_index(name='Quantidade')
sobreviventes
```

11 - Quantas mulheres não sobreviveram?

```python
#Filtra na coluna 'Sobreviveu' apenas os valores 'Sim'. Agrupa os mesmos pela coluna Sexo, conta as ocorrências e reinicia o índice e renomeia a coluna
sexo_sobreviventes = data[data['Sobreviveu'] == 'Sim'].groupby('Sexo')[['Sobreviveu']].count().reset_index()
sexo_sobreviventes.columns = ['Sexo', 'Quantidade']

print(sexo_sobreviventes)
```

12 - Proporcionalmente, sobreviveram mais homens ou mais mulheres? Cite as proporções.

```python
#Filtra os valores da coluna 'Sobreviveu' com os valores 'Sim' e calcula as proporções das ocorrências em porcentagem
proporcao_sexo = data[data['Sobreviveu'] == 'Sim']['Sexo'].value_counts(normalize=True) * 100

print(round(proporcao_sexo,2))
```

13 - Levando-se em consideração a idade dos passageiros, qual a idade e quantidade de pessoas com o maior número de mortos?

```python
# Filtra o DataFrame para incluir apenas os passageiros que não sobreviveram, conta os valores e obtem os rótulos do indice
mortos_idade = data[data['Sobreviveu'] == 'Não']['Idade'].value_counts().index[0]

#Filtra as ocorrencias onde a coluna 'Sobreviveu' possui valor 'Não' e a idade igual a idade de 'mortos_idade'
#.shape[0] obtem as dimensões do DataFrame resultante. [0] (linhas) retorna a quantidade de passageiros com a idade mais comum
mortos_qtd = data[(data['Sobreviveu'] == 'Não') & (data['Idade'] == mortos_idade)].shape[0]

print(f'A idade com maior registro de mortos é: {mortos_idade}')
print(f'A quantidade de mortos é: {mortos_qtd}')
```

14 - Qual a média de idade dos homens sobreviventes?

```python
#Filtra apenas os homens que sobreviveram, agrupa os valores filtrados pela coluna 'Sexo', calcula a média da idade dos mesmos.
media_idade = data[(data['Sexo'] == 'Homem') & (data['Sobreviveu'] == 'Sim')].groupby('Sexo').mean().reset_index()

#Exibe apenas as colunas 'Sexo' e 'Idade'
print(media_idade[['Sexo', 'Idade']])
```

15 - Levando-se em consideração passageiros prioritários (mulheres e crianças de até 15 anos independente do sexo) qual a proporção de sobreviventes por sexo?

```python
#Filtra os passageiros com idade maior que 0 e menores ou iguais a 15 ou com o 'sexo' igual a mulher
filtro = (data['Idade'] <= 15) & (data['Idade'] > 0) | (data['Sexo'] == 'Mulher')

#Filtra o DataFrame original com base no filtro estabelecido
filtrado = data[filtro]

#Substitui os valores da coluna 'Sobreviveu' com os valores 'Sim' e 'Não'
filtrado['Sobreviveu'] = filtrado['Sobreviveu'].replace(['Sim','Não'],[1,0])

#Realiza o agrupamento dos valores filtrados através da coluna 'Sexo' e calcula a proporção por sexo
proporcao_criancas_mulheres = filtrado.groupby('Sexo')['Sobreviveu'].mean() * 100

#Exibe a proporção
print(proporcao_criancas_mulheres)
```

18 - Crie um dataframe que demonstre a quantidade de sobreviventes e não sobreviventes, agrupados por sexo e classe.

```python
#Substitui os valores da coluna 'Classe' como 1 = A / 2 = B / 3 = C
data['Classe']= data['Classe'].replace([1, 2, 3],['A','B','C'])

#Substitui os valores da coluna 'Sobreviveu' como 0 = Não e 1 = Sim
data['Sobreviveu']= data['Sobreviveu'].replace([0, 1],['Não','Sim'])

# Cria uma tabela dinâmica contando a quantidade de sobreviventes e não sobreviventes para cada sexo e classe
data.pivot_table('Sobreviveu', ['Classe'], 'Sexo', aggfunc='count')

# Cria uma tabela dinâmica contando a quantidade de sobreviventes e não sobreviventes para cada combinação de sexo e classe,
# preenchendo com 0 onde não há dados, usando 'size' como a função de agregação
data.pivot_table(index=['Sexo', 'Classe'], columns='Sobreviveu', aggfunc='size', fill_value=0)
```

19 - Dos homens com idade entre 24 e 30 anos quantos da classe 3 sobreviveram? Quantos da classe 2 não sobreviveram?

```python
#Filtra o DataFrame com apenas homens da classe C, com idade entre 24 e 30 anos e sobreviventes
filtro = data[(data['Sexo'] == 'Homem') & (data['Classe'] == 'C') & (data['Idade'].between(24, 30)) & (data['Sobreviveu'] == 'Sim')]

#Cria uma tabela dinâmica para contar a quantidade de sobreviventes
tabela_contagem = filtro.pivot_table('Sobreviveu', ['Classe'], 'Sexo', aggfunc='count')

# Cria uma tabela dinâmica com os devidos filtros, preenchendo com 0 onde não há dados, usando 'size' como a função de agregação
tabela_din = filtro.pivot_table(index=['Sexo', 'Classe'], columns='Sobreviveu', aggfunc='size', fill_value=0)

#Exibe a tabela criada
tabela_din
```
